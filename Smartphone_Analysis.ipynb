**Dataset is containing null values so we need to fill null values from some desire number**

```python
#Filling null values by "no"
df["Fingerprints"] = df["Fingerprints"].fillna("No")
```

```python
#Filling null values by "no"
df["5G"] = df["5G"].fillna("No")
```

```python
#filling null values by "60.0"
df["Refresh_Rate(hz)"] = df["Refresh_Rate(hz)"].fillna("60.0")
```

***I have filled null column of Fingerprint and 5G column with "No" because all of phone_name that have mentioned in front of null column that are low
budge phones and old model phone so low budget phone don't provide fingerprint feature and old model didn't have 5G.***

### Brand_Name

```python
plt.figure(figsize=(28, 6))
ax = sns.countplot(x = "Brand_Name", data = df)
ax.bar_label(ax.containers[0])

plt.xticks(rotation = 45)
plt.show()
```

**Number of units sale by each Brand Name**

```python
df.groupby(['Brand_Name'],as_index = False)['Price'].sum().sort_values(by = 'Price' ,ascending = False)
```

**Total Revenue generated by each Brand** 

```python
phone_sales = df.groupby(['Brand_Name'],as_index = False)['Price'].sum().sort_values(by = 'Price' ,ascending = False).head(12)

plt.figure(figsize=(20, 6))
sns.barplot(x = 'Brand_Name', y = 'Price', data = phone_sales)
```
